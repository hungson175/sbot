# 01 — Generators: Iterators Without the Boilerplate

**Goal:** Understand `yield` deeply — generators are the engine that powers async/await.

## The insight

Building `MyRange` in the previous notebook required TWO classes and ~20 lines.
A generator does the same thing:

```python
def my_range(n):
    i = 0
    while i < n:
        yield i    # pause here, give caller the value, resume later
        i += 1
```

A function with `yield` doesn't run when called — it returns a **generator object** that IS an iterator.

## Summary

```
Iterator: __next__() → value → __next__() → value → StopIteration
Generator: next()    → yield  → next()    → yield  → return
Coroutine: resume    → await  → resume    → await  → return
```

Same pattern. Just different names.
