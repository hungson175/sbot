## Summary

| Primitive | What it does | Use case |
|-----------|-------------|----------|
| `Lock` | Mutual exclusion (1 thread) | Protect shared variable |
| `RLock` | Reentrant lock | Methods calling other locked methods |
| `Event` | Signal between threads | "Server is ready" notification |
| `Semaphore` | Allow N concurrent accesses | Connection pool, rate limiting |
